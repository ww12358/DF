```python
import pandas as pd


dtypes = {
        'Date_time':'datetime64[ns]',
        'st_dn' :   'int64',
        'sw_dn' :   'int64',
        'tr_dn' :   'int64',
        'st'    :   'int64',
        'sw'    :   'int64',
        'tr'    :   'int64',
        'total' :   'int64',
        'st_pct':   'float64',
        'sw_pct':   'float64',
        'tr_pct':   'float64',
        'tt_pct':   'float64',
        'position': 'float64'
          }



symbol_collection=[
                   'A','M','Y', 'P', 'C','JD',
                   'SR', 'CF', 'WH', 'OI', 'RM',  'RS', 'RI',
                   'CU', 'AL', 'ZN', 'PB', 'RU', 'AU', 'AG', 'RB',
                   'TA', 'ME', 'ZC', 'FG',
                   'PP', 'L', 'V', 'I', 'J', 'JM'
                   'IF', 'IH', 'IC', 'TF'
                   ]
headers = ['Date_time', 'st_dn','sw_dn','tr_dn','st','sw','tr','total','st_pct','sw_pct','tr_pct','tt_pct','position']
```

```python
symbol = 'RB'

xl = pd.ExcelFile('/home/sean/Nutstore/DY强弱记录(2018).xlsm')
f = pd.HDFStore('./data/creekIDX.hdf5')

try:
    df = pd.read_excel(xl, sheet_name=symbol, names=headers)
    dfo = pd.read_hdf(f, '/' + symbol, mode='r')
except:
    print 'Error opening excel file'
df.drop(df.index[0], inplace=True)
#diff_pd(df, dfo)
print df
#print df.dtypes
df = df.astype(dtype=dtypes)
print df.dtypes
```

```python
i1 = df.set_index('Date_time').index
i2 = dfo.set_index('Date_time').index

df_new = df[~i1.isin(i2)]
dfo.update(df)

dfc = dfo.append(df_new)

dfc.reset_index(inplace=True, drop = True)

dfc.to_hdf(f, '/'+symbol, format='table', append=True, data_columns=True, mode='w')

f.flush()
f.close()
```
